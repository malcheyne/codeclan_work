```python
import pandas as pd

demo_data = pd.read_csv("data/life_expectancy_and_income.csv")
```

```python
demo_data_2019 = demo_data.loc[demo_data.year == 2019].copy()

demo_data_2019
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [5, 3]
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.size'] = 10
```

```python
plt.scatter(x = "income_per_person", 
            y = "life_expectancy", 
            data = demo_data_2019)
```

```python
plt.scatter(x = "income_per_person", 
            y = "life_expectancy", 
            data = demo_data_2019)
plt.xlabel("Income per Person ($)")
plt.ylabel("Life Expectancy")
plt.title("Scatterplot of Income per Person vs. Life Expectancy")
```

```python
demo_data_2019.loc[:, "income_per_person_1000"] = demo_data_2019.income_per_person/1000
```

```python
demo_data_2019.head()
```

```python
plt.scatter(x = "income_per_person_1000",
            y = "life_expectancy",
            data = demo_data_2019)
plt.xlabel('Income per person ($1000, ppp)') 
plt.ylabel('Life expectancy') 
plt.title("Scatterplot of Income per person vs. life expectancy")
```

#### Setting limits

We may want to 'zoom in' on a particular part of the graph to see it in more detail. We can do this "zooming" with the `plt.xlim` and the `plt.ylim` functions. Below we can see an example of zooming in on the x-axis.

Let's say we are particularly interested in an income between $0 and \$10,000 ppp because there is a lot of overlapping points here.


```python
plt.scatter(x = "income_per_person", 
            y = "life_expectancy", 
            data = demo_data_2019)
plt.xlim(0, 10000)
```

#### Axis ticks

You may wish to customise the tick marks on your graph, for example they may be overlapping and you want to rotate them or perhaps you want to specify your own tick marks.

For example let's say we only want our `life_expectancy` axis to be every 10 years, rather than the default of every 5. We can use the `plt.yticks` for this.


```python
plt.scatter(x = "income_per_person", 
            y = "life_expectancy", 
            data = demo_data_2019)
plt.yticks([50, 60, 70 ,80]);
# use ; to suppress output of return - try without ; to see output
```

```python
plt.scatter(x = "income_per_person", 
            y = "life_expectancy", 
            data = demo_data_2019)
plt.yticks([50, 60, 70 ,80])
```

### Using Seaborn

So let's begin using seaborn. First thing to do is we need to import it, conventionally as `sns`:

```python
import seaborn as sns
```

```python
sns.scatterplot(x = "income_per_person", 
                y = "life_expectancy", 
                data = demo_data_2019);
```

```python
sns.histplot(x = "income_per_person", 
             data = demo_data_2019)
```

```python
sns.boxplot(x = "income_per_person", 
            data = demo_data_2019);
```

```python

demo_data_japan = (
    demo_data
    .loc[demo_data['country'] == 'Japan']
    .copy()
)
```

```python
sns.lineplot(x = "year", 
             y = "fertility_rate", 
             data = demo_data_japan)
plt.title("Fertility rate over time for Japan")
```

```python
sns.barplot(x = 'country',
           y = 'life_expectancy',
           data = demo_data_2019
           .head(5)
           )
```

```python
ordered_data = (
    demo_data_2019.head(5)
    .sort_values('life_expectancy')['country']
    )

sns.barplot(x = 'country',
           y = 'life_expectancy',
           data = demo_data_2019
           .head(5),
            order = ordered_data
           )
```

```python
last_20 = (
    demo_data
    .loc[demo_data['year'] > 2000]
    .copy()
    )
```

```python
import numpy as np
```

```python
sns.barplot(x = "year", 
            y = "income_per_person", 
            data = last_20, 
            estimator = np.mean)
```

```python
sns.barplot(x = "year", 
            y = "income_per_person", 
            data = last_20, 
            estimator = np.mean, 
            ci = False)
plt.title("Average income per person by year from 2000 to 2019")
```

```python
sns.barplot(x = "year", 
            y = "income_per_person", 
            data = last_20, 
            estimator = np.median, 
            ci = False)
plt.title("Median income per person by year from 2000 to 2019")
```

```python
sns.barplot(x = "income_per_person", 
            y = "year", 
            data = last_20, 
            estimator = np.mean, 
            ci = False, 
            orient = "h")
plt.title("Average income per person by year from 2000 to 2019")
```

```python
(sns.barplot(x = "income_per_person", 
            y = "year", 
            data = last_20, 
            estimator = np.mean, 
            ci = False, 
            orient = "h")
            .invert_yaxis)
plt.title("Average income per person by year from 2000 to 2019")
```

```python
sns.barplot(x = "year", 
            y = "income_per_person", 
            data = last_20, 
            estimator = np.median, 
            ci = False)
plt.title("Median income per person by year from 2000 to 2019")
plt.xticks(rotation = 45);
```
