### Sorting data

You can sort data with the `sort_values()` method

```python
demographic_data.sort_values("life_expectancy")
```

```python
demographic_data.sort_values("life_expectancy", ascending = False)
```

To persist the change, we either need to overwrite the data frame, as we saw earlier, or we can use the `inplace = True` argument

```python
demographic_data.sort_values("life_expectancy", 
                             ascending = False, 
                             inplace = True)
demographic_data.head()
```

### Creating and changing columns
 
#### Creating columns

You can create a new column using a similar syntax to selecting a column. Here we want to create a new column called `years_since_1900` which tells us how many years since 1900 for the year entry.

```python
demographic_data['years_since_1900'] =demographic_data['year'] - 1900
```

```python
demographic_data
```

#### Changing columns

You can use the same method to change an existing column. Below we are changing income from being measured in dollars to pounds (roughly!). On the left hand side again we use `loc[]` to specify which column we are changing.

```python
demographic_data.loc[:, "income_per_person"] = demographic_data[
    "income_per_person"]*0.8
    
#demographic_data.head()
demographic_data.tail()
```
