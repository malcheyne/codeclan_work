#### Calculating summaries in groups

Knowing these summary statistics, such a mean, max and min, for each level of a categorical variable can be hugely insightful to see how categories compare. For example let's say we wanted to know the average life expectancy for each country across the years of data in our dataframe. We could go through each country in turn to calculate this, starting with 'Afghanistan' then 'Albania' and so on.

We might be here a while to get through all these! Luckily for us Python has a handy function `groupby()` that allows us to chain other methods to it, like `min()`, `max()`, or `mean()` to get insight from each category level easily.

There is two ways to do a group by - either on the whole data, or on a specific column. First, lets see an example with the whole data

```python
demographic_data.groupby("country").mean()
```

```python
demographic_data.groupby("country")['income_per_person'].mean().reset_index(name = 'mean_income')
```

```python
(
    demographic_data
    .groupby("country")['income_per_person']
    .mean()
    .reset_index(name = 'mean_income')
)
```

```python
demographic_data.groupby("country")['income_per_person']\
.mean().reset_index(name = 'mean_income')
```

```python
demographic_data.groupby("country")['income_per_person'].mean(
).reset_index(name = 'mean_income')
```

### Missing values

In R, if you try to calculate the mean of a column that contains a missing value the answer will be a missing value. However, in Python the same behaviour you see when you set `na.rm = TRUE` happens by default.

Let's load in the `demographic_data` again but this time it has some missing values.


```python
demographic_data_missing = (
    pd.read_csv(
        "data/life_expectancy_and_income_missing.csv"
    )
)
```

first bit needs to be outside of the brackets. We\'re assigning the output of the method chain (everything in the brackets) to the object `demographic_data` so demographic data shouldn\'t be inside the brackets. 

```python
demographic_data_missing = pd.read_csv("data/life_expectancy_and_income_missing.csv")
```

```python
demographic_data_missing.head()
```

```python
demographic_data_missing.isna().sum()
```

```python
demographic_data_missing.fillna(0)
```

```python
demographic_data_missing.fillna({
  "life_expectancy": demographic_data_missing["life_expectancy"].mean(),
  "income_per_person": demographic_data_missing["income_per_person"].mean()
})
```

```python
(
    demographic_data_missing
    .groupby('country')['life_expectancy']
    .transform("mean")
)
```

```python
demographic_data_missing.fillna({
  "life_expectancy": demographic_data_missing
    .groupby("country")["life_expectancy"]
    .transform("mean")
})
```

```python
demographic_data_missing.fillna({
  "life_expectancy": demographic_data_missing
    .groupby("country")["life_expectancy"]
    .transform("mean")
    .map(round)
})
```

```python
(
    demographic_data_missing
    .dropna()
)
```
